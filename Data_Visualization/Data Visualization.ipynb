# 1. Bar Chart

A bar chart can be used to compare the population of different cities.

| City        | Population (in millions) |
|-------------|--------------------------|
| City A      | 5.3                      |
| City B      | 3.8                      |
| City C      | 4.1                      |
| City D      | 2.7                      |
| City E      | 3.5                      |

```python
import matplotlib.pyplot as plt  # Import the matplotlib library for plotting

# Data
cities = ['City A', 'City B', 'City C', 'City D', 'City E']
population = [5.3, 3.8, 4.1, 2.7, 3.5]

# Create a bar chart
plt.figure(figsize=(10, 6))  # Create a new figure with specified size (width=10 inches, height=6 inches)
bars = plt.bar(cities, population, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'])
# Create a bar chart with cities on x-axis, population on y-axis, using custom colors for each bar

# Customize labels and title
plt.xlabel('City', fontdict={'fontsize': 14, 'fontweight': 'bold', 'family': 'serif'})
# Set label for x-axis with custom font properties: fontsize=14, bold, serif font family
plt.ylabel('Population (in millions)', fontdict={'fontsize': 14, 'fontweight': 'bold', 'family': 'serif'})
# Set label for y-axis with custom font properties: fontsize=14, bold, serif font family
plt.title('Population of Different Cities', fontdict={'fontsize': 18, 'fontweight': 'bold', 'family': 'serif'})
# Set title for the plot with custom font properties: fontsize=18, bold, serif font family

# Rotate x-axis labels for better readability and customize font properties
plt.xticks(rotation=45, fontname='serif', fontsize=12)  
# Rotate x-axis labels by 45 degrees, use serif font for tick labels, and set fontsize=12 for tick labels

# Annotate each bar with its value
for bar in bars:
    yval = bar.get_height()  # Retrieve the height of the bar
    plt.text(bar.get_x() + bar.get_width()/2, yval, format(yval, ','),  # Position and format of the text
             va='bottom', ha='center', fontsize=10, fontweight='bold', family='serif')  # Text properties

# Display the chart
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()  # Display the plot
```

# 4. Heat Map

A heat map can be used to show the density of people in different areas of a city.

| Area       | Density (people per sq km) |
|------------|-----------------------------|
| Area 1     | 1000                        |
| Area 2     | 1500                        |
| Area 3     | 800                         |
| Area 4     | 2000                        |
| Area 5     | 1800                        |

```python
import matplotlib.pyplot as plt  # Import the matplotlib library for plotting
import numpy as np  # Import numpy for array manipulation

# Data
areas = ['Area 1', 'Area 2', 'Area 3', 'Area 4', 'Area 5']
density = [1000, 1500, 800, 2000, 1800]

# Convert density data to a 1x5 matrix for heatmap representation
data = np.array(density).reshape(1, -1)  # Reshape density list into a 1x5 numpy array

# Create a heatmap
plt.figure(figsize=(8, 3))  # Create a new figure with specified size (width=8 inches, height=3 inches)
heatmap = plt.imshow(data, cmap='YlOrRd', aspect='auto')
# Plot the heatmap with data, using 'YlOrRd' colormap, and auto aspect ratio

# Customize labels and title
plt.xticks(np.arange(len(areas)), areas, rotation=45, ha='right')
# Set x-axis ticks at positions and labels for each area, rotate labels by 45 degrees, align them to the right
plt.yticks([])  # Disable y-axis ticks (not needed for a single row heatmap)
plt.xlabel('Area', fontsize=12, fontweight='bold', family='serif')
# Set label for x-axis with custom font properties: fontsize=12, bold, serif font family
plt.title('Density of People in Different Areas of a City', fontsize=14, fontweight='bold', family='serif')
# Set title for the plot with custom font properties: fontsize=14, bold, serif font family

# Display the color bar
plt.colorbar(heatmap)  # Display color bar indicating the density scale

# Display the heatmap
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()  # Display the plot
```

### 7. Bubble Chart

A bubble chart can be used to represent sales data where the bubble size represents the profit and the color represents the region.

| Sales | Profit | Region  |
|-------|--------|---------|
| 1000  | 200    | North   |
| 1500  | 300    | South   |
| 1200  | 250    | East    |
| 1300  | 400    | West    |
| 1100  | 220    | Central |

```python
import matplotlib.pyplot as plt  # Import the matplotlib library for plotting

# Data
sales = [1000, 1500, 1200, 1300, 1100]
profit = [200, 300, 250, 400, 220]
regions = ['North', 'South', 'East', 'West', 'Central']

# Scale profit values for bubble size (optional)
scaled_profit = [p * 2 for p in profit]  # Scale profit values for better visualization

# Define colors for regions
colors = {'North': 'blue', 'South': 'green', 'East': 'orange', 'West': 'red', 'Central': 'purple'}

# Create a bubble chart
plt.figure(figsize=(10, 6))  # Create a new figure with specified size (width=10 inches, height=6 inches)
for i in range(len(sales)):
    plt.scatter(sales[i], profit[i], s=scaled_profit[i], c=colors[regions[i]], alpha=0.6, edgecolors='w', linewidth=2)
# Plot a scatter plot for each data point, with sales on x-axis, profit on y-axis, bubble size scaled by profit, color based on region,
# transparency of bubbles set to 0.6, white edge around bubbles, and edge line width set to 2

# Add labels and title
plt.xlabel('Sales', fontsize=14, fontweight='bold', family='serif')
# Set label for x-axis with custom font properties: fontsize=14, bold, serif font family
plt.ylabel('Profit', fontsize=14, fontweight='bold', family='serif')
# Set label for y-axis with custom font properties: fontsize=14, bold, serif font family
plt.title('Bubble Chart: Sales vs. Profit (by Region)', fontsize=16, fontweight='bold', family='serif')
# Set title for the plot with custom font properties: fontsize=16, bold, serif font family

# Create a legend
legend_labels = [plt.scatter([], [], s=scaled_profit[i], c=colors[regions[i]], edgecolors='w', linewidth=2) for i in range(len(sales))]
# Create empty scatter plots with bubble size scaled by profit, colored by region, with white edge and line width 2
plt.legend(legend_labels, regions, loc='upper left', title='Region', title_fontsize='medium', fontsize='medium', markerscale=0.5)
# Display legend with labels for each region, located at upper left, with title 'Region', title font size 'medium', and label font size 'medium', marker scale 0.5

# Display the chart
plt.grid(True)  # Display grid lines on the plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()  # Display the plot
```

### 8. Funnel Chart

A funnel chart can be used to illustrate the sales process stages.

| Stage          | Number of Leads |
|----------------|------------------|
| Prospects      | 1000             |
| Qualified Leads| 700              |
| Proposals Sent | 500              |
| Negotiations   | 300              |
| Closed Deals   | 150              |

```python
import matplotlib.pyplot as plt  # Import the matplotlib library for plotting

# Data
stages = ['Prospects', 'Qualified Leads', 'Proposals Sent', 'Negotiations', 'Closed Deals']
leads = [1000, 700, 500, 300, 150]

# Calculate the width of each stage (simulate funnel shape)
widths = [leads[0]] + [leads[i] - leads[i+1] for i in range(len(leads)-1)]
# Calculate the width of each horizontal bar, where each subsequent bar is smaller than the previous one

# Create a funnel chart
plt.figure(figsize=(8, 6))  # Create a new figure with specified size (width=8 inches, height=6 inches)
plt.barh(stages, widths, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'])
# Create a horizontal bar chart with stages on y-axis, widths as bar lengths, using custom colors for each bar

# Add labels and title
plt.xlabel('Number of Leads', fontsize=14, fontweight='bold', family='serif')
# Set label for x-axis with custom font properties: fontsize=14, bold, serif font family
plt.ylabel('Sales Process Stages', fontsize=14, fontweight='bold', family='serif')
# Set label for y-axis with custom font properties: fontsize=14, bold, serif font family
plt.title('Funnel Chart: Sales Process Stages', fontsize=16, fontweight='bold', family='serif')
# Set title for the plot with custom font properties: fontsize=16, bold, serif font family

# Display the chart
plt.grid(True)  # Display grid lines on the plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()  # Display the plot
```

### 9. Radar Chart

A radar chart can be used to analyze the performance of different departments based on various criteria.

| Department | Quality | Efficiency | Innovation | Customer Satisfaction |
|------------|---------|------------|------------|-----------------------|
| Sales      | 8       | 7          | 6          | 9                     |
| Marketing  | 7       | 6          | 8          | 8                     |
| IT         | 9       | 8          | 7          | 7                     |
| HR         | 6       | 7          | 5          | 8                     |

```python
import matplotlib.pyplot as plt  # Importing the matplotlib library for plotting

# Data for departments and their metrics
departments = ['Sales', 'Marketing', 'IT', 'HR']  # List of department names
qualities = [8, 7, 9, 6]  # List of quality scores for each department
efficiencies = [7, 6, 8, 7]  # List of efficiency scores for each department
innovations = [6, 8, 7, 5]  # List of innovation scores for each department
customer_satisfaction = [9, 8, 7, 8]  # List of customer satisfaction scores for each department

# Position of each department on x-axis
x = range(len(departments))  # Creating a range of indices for the departments

# Width of the bars
width = 0.1  # Setting the width of each bar in the column chart

# Plotting the bars for each metric
plt.bar(x, qualities, width=width, label='Quality')  # Plotting bars for quality scores
plt.bar([pos + width for pos in x], efficiencies, width=width, label='Efficiency')  # Plotting bars for efficiency scores
plt.bar([pos + 2 * width for pos in x], innovations, width=width, label='Innovation')  # Plotting bars for innovation scores
plt.bar([pos + 3 * width for pos in x], customer_satisfaction, width=width, label='Customer Satisfaction')  # Plotting bars for customer satisfaction scores

# Adding labels, title, and legend
plt.xlabel('Departments')  # Setting the label for the x-axis
plt.ylabel('Scores')  # Setting the label for the y-axis
plt.title('Departmental Performance Metrics')  # Setting the title of the plot
plt.xticks([pos + 1.5 * width for pos in x], departments)  # Setting the positions and labels of the x-axis ticks
plt.legend()  # Adding a legend to the plot

# Showing the plot
plt.tight_layout()  # Adjusting subplot parameters to give specified padding
plt.show()  # Displaying the plot
```

### 10. Gauge

A gauge chart can be used to display the percentage of a sales target achieved.

| Sales Target Achieved |
|-----------------------|
| 75%                   |

```python
import matplotlib.pyplot as plt  # Importing matplotlib library for plotting
import numpy as np  # Importing numpy for numerical operations

# Data
target_achieved = 75  # Percentage of sales target achieved

# Create a figure and axis
fig, ax = plt.subplots(figsize=(6, 6))  # Creating a figure and axis with a square size

# Plot a gauge-like chart
ax.set_xlim([-1, 1])  # Setting the x-axis limits to create a square plotting area
ax.set_ylim([-1, 1])  # Setting the y-axis limits to create a square plotting area

# Draw a colored arc representing the percentage achieved
theta = np.linspace(0, 2 * np.pi, 100)  # Creating an array of angles from 0 to 2π (360 degrees) with 100 points
ax.fill_betweenx(np.sin(theta), np.cos(theta), where=(theta <= (2 * np.pi * target_achieved / 100)), color='green', alpha=0.5)  # Filling the area with green up to the achieved percentage
ax.fill_betweenx(np.sin(theta), np.cos(theta), where=(theta > (2 * np.pi * target_achieved / 100)), color='red', alpha=0.5)  # Filling the remaining area with red

# Draw a white circle in the middle to represent the gauge
circle = plt.Circle((0, 0), 0.6, color='white', ec='black')  # Creating a white circle with a black edge
ax.add_artist(circle)  # Adding the circle to the plot

# Add a text label showing the percentage achieved
ax.text(0, 0, f'{target_achieved}%', fontsize=50, ha='center', va='center')  # Adding a text label at the center showing the achieved percentage

# Remove grid and ticks
ax.grid(False)  # Turning off the grid
ax.set_xticks([])  # Removing x-axis ticks
ax.set_yticks([])  # Removing y-axis ticks

# Add a title
ax.set_title('Sales Target Achieved')  # Setting a title for the plot

# Show the plot
plt.show()  # Displaying the plot
```

### 12. Marimekko Chart

A Marimekko chart can be used to show market share distribution across different regions and product categories.

| Region | Product A | Product B | Product C |
|--------|-----------|-----------|-----------|
| North  | 20%       | 30%       | 10%       |
| South  | 10%       | 20%       | 40%       |
| East   | 30%       | 10%       | 10%       |
| West   | 10%       | 10%       | 30%       |


```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the Marimekko chart
regions = ['North', 'South', 'East', 'West']
product_a = [20, 10, 30, 10]
product_b = [30, 20, 10, 10]
product_c = [10, 40, 10, 30]

# Convert percentages to proportions
total = np.array(product_a) + np.array(product_b) + np.array(product_c)
product_a = np.array(product_a) / total
product_b = np.array(product_b) / total
product_c = np.array(product_c) / total

# Calculate cumulative sums for stacked bar positions
cumulative_a = product_a
cumulative_b = cumulative_a + product_b
cumulative_c = cumulative_b + product_c

# Plotting the Marimekko chart
fig, ax = plt.subplots(figsize=(10, 6))  # Creating a figure and axis for plotting

# Plotting each product as a stacked bar
ax.bar(regions, product_a, color='skyblue', label='Product A')
ax.bar(regions, product_b, bottom=product_a, color='salmon', label='Product B')
ax.bar(regions, product_c, bottom=cumulative_b, color='lightgreen', label='Product C')

# Adding labels and title
ax.set_xlabel('Region')  # Setting label for x-axis
ax.set_ylabel('Proportion')  # Setting label for y-axis
ax.set_title('Market Share Distribution by Region and Product')  # Setting title for the plot

# Adding legend
ax.legend()  # Adding a legend to the plot

# Displaying the plot
plt.show()  # Showing the generated Marimekko chart
```

### 13. Donut Chart

A donut chart can be used to show the distribution of expenses.

| Expense       | Percentage |
|---------------|------------|
| Rent          | 30%        |
| Salaries      | 40%        |
| Utilities     | 10%        |
| Marketing     | 15%        |
| Miscellaneous | 5%         |

```python
import matplotlib.pyplot as plt  # Importing matplotlib library for plotting

# Data for the donut chart
expenses = ['Rent', 'Salaries', 'Utilities', 'Marketing', 'Miscellaneous']  # Expense categories
percentages = [30, 40, 10, 15, 5]  # Corresponding percentages for each expense

# Plotting the donut chart
fig, ax = plt.subplots(figsize=(8, 8))  # Creating a figure and axis for plotting with specified size
wedges, texts, autotexts = ax.pie(percentages, labels=expenses, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4))  # Plotting the pie chart with specified properties

# Adding a circle at the center to create a donut shape
centre_circle = plt.Circle((0, 0), 0.70, fc='white')  # Creating a white circle to place at the center
fig.gca().add_artist(centre_circle)  # Adding the circle to the plot

# Adding title to the plot
ax.set_title('Expense Distribution')  # Setting title for the plot

# Displaying the plot
plt.show()  # Showing the generated donut chart
```

# 14. Area Chart

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd

# Create example data using a dictionary
data = {
    'Year': [2015, 2016, 2017, 2018, 2019, 2020],
    'Category A': [3, 4, 6, 7, 8, 9],
    'Category B': [2, 3, 5, 6, 7, 8],
    'Category C': [1, 2, 3, 4, 5, 6]
}

# Convert dictionary to a DataFrame
df = pd.DataFrame(data)

# Set the 'Year' column as the index
df.set_index('Year', inplace=True)

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the area chart
# 'df.plot.area' plots an area chart for all columns in the DataFrame
df.plot.area(ax=ax, alpha=0.5)

# Set chart title and labels
ax.set_title('Stacked Area Chart Example')  # Title of the chart
ax.set_xlabel('Year')  # Label for the x-axis
ax.set_ylabel('Values')  # Label for the y-axis

# Customize the x and y axis ticks
ax.set_xticks(df.index)  # Set x-axis ticks to match the 'Year' index
ax.set_yticks(range(0, 21, 2))  # Set y-axis ticks from 0 to 20 with step of 2

# Add a legend
ax.legend(title='Categories')  # Add a title to the legend

# Display the plot
plt.show()
```
